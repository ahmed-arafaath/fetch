## Receipts

```python
# Import necessary libraries and upload receipts data
import pandas as pd
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Geting the file name


receipts_df = pd.read_csv(filename)
receipts_df.head()
```

```python
receipts_df.describe()
```

```python
receipts_df.info()
```

```python
# Assumption: receipt_id should be unique in this table
receipts_df['receipt_id'].nunique() # this concludes that there are no duplicates in the table as well unless there same values are populated for different ids
```

```python
# Assumption: Total spent will not be a negative value since this table will only have order receipts and not the return receipts
receipts_df['totalSpent'].min()
```

Rewards funnel from a user journey perspective (left to right):

purchase date >> create date (purchasing leads to creating event) >> date scanned >> points awarded date >> finished date

Modify date can fall anywhere after the creat date in the funnle

```python
# Convert date type for all dates
receipts_df['purchaseDate'] = pd.to_datetime(receipts_df['purchaseDate'], errors='coerce')
receipts_df['createDate'] = pd.to_datetime(receipts_df['createDate'], errors='coerce')
receipts_df['dateScanned'] = pd.to_datetime(receipts_df['dateScanned'], errors='coerce')
receipts_df['finishedDate'] = pd.to_datetime(receipts_df['finishedDate'], errors='coerce')
receipts_df['pointsAwardedDate'] = pd.to_datetime(receipts_df['pointsAwardedDate'], errors='coerce')
receipts_df['modifyDate'] = pd.to_datetime(receipts_df['modifyDate'], errors='coerce')
```

```python
receipts_df.info()
```



1.   PurchaseDate - 671
2.   CreateDate - 1119
3.   dateScanned - 1119
4.   pointsAwardedDate - 537
5.   finishedDate - 568

**Every receipt_id will involve a purchaseDate, in this case we are missing records for almost half of the receipts data.**



```python
validation_df = receipts_df.copy()
validation_df['pruchase_create'] = receipts_df['purchaseDate'] > receipts_df['createDate'] # returns a bool
validation_df['pruchase_create'].sum() # only counts when the value is True
```

There were 13 cases where purchase date is greater than create date which is an data inconsistency issue

```python
validation_df['create_scanned'] = receipts_df['createDate'] > receipts_df['dateScanned']
validation_df['create_scanned'].sum()
# this is valid
```

```python
validation_df['scanned_pointsawarded'] = receipts_df['dateScanned'] > receipts_df['pointsAwardedDate']
validation_df['scanned_pointsawarded'].sum()
# This is also valid
```

```python
validation_df['pointsawarded_finished'] = receipts_df['pointsAwardedDate'] > receipts_df['finishedDate']
validation_df['pointsawarded_finished'].sum()
# This is also valid
```

```python
validation_df['finished_create'] = receipts_df['createDate'] > receipts_df['pointsAwardedDate']
validation_df['finished_create'].sum()
# This is also valid
```

```python
validation_df['finished_purchase'] = receipts_df['purchaseDate'] > receipts_df['pointsAwardedDate']
validation_df['finished_purchase'].sum()
# This is not valid it says points are awarded to users even before a user made their purchase
```

```python
validation_df['finished_purchase'] = receipts_df['purchaseDate'] > receipts_df['finishedDate']
validation_df['finished_purchase'].sum()
# This is not valid it says process is finisged even before a user made their purchase
```

```python
# Unusual Frequency of Records

validation_df['receipt_count'] = receipts_df.groupby('user_id')['receipt_id'].transform('count')
unusual_frequency = validation_df[validation_df['receipt_count'] > validation_df['receipt_count'].quantile(0.99)]
print("Unusual frequency of receipts per user:\n", unusual_frequency)
```

While null values are common in the real world, there are some set busines processes/sequence the data must follow in order for Analyst to make inference from them.

Summary of the preliminary inconsistency check,
1. Every receipt should ideally have purchase date, total spent (assuming the table will only have order receipts and not the return receipts), purchasedItemCount, user_id
2. Based on the rewards funnel, purchase date should always be lesser than create date or records in later stage of the funnel, in this case there were 13 anomalies.
